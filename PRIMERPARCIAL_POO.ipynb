10. Explica el siguiente código:

 df = pd.DataFrame({'key1' : ['a', 'a', 'b', 'b', 'a'], 'key2' : ['one', 'two', 'one', 'two', 'one'], 'data1' : np.random.randn(5),'data2' : np.random.randn(5)})

  (a) Quita la columna de "key 2".

  (b) Cambia el índice a tu conveniencia.

  (c) Realiza un análisis de estadística descriptiva (No es necesario explicar cada resultado).

  (d) Calcula la media de cada columna sobrante.

```python
# Crear un DataFrame 'df' con cuatro columnas 'key1', 'key2', 'data1' y 'data2' usando pandas
df = pd.DataFrame({'key1': ['a', 'a', 'b', 'b', 'a'], 'key2': ['one', 'two', 'one', 'two', 'one'], 'data1': np.random.randn(5), 'data2': np.random.randn(5)})

# Mostrar el DataFrame 'df'
df
```

```python
# Crear un nuevo DataFrame 'dfa' eliminando la columna 'key2' de 'df'
dfa = df.drop(columns='key2')

# Mostrar el DataFrame 'dfa' resultante
print(dfa)
```

```python
# Establecer la columna 'key1' como índice en el DataFrame 'df'
df = df.set_index('key1')

# Mostrar el DataFrame 'df' con 'key1' como índice
print(df)
```

```python
# Calcular estadísticas descriptivas del DataFrame 'df' y guardarlas en 'desc'
desc = df.describe()

# Mostrar las estadísticas descriptivas
print(desc)
```

```python
# Calcular el promedio de la columna 'data1' en el DataFrame 'df' y asignarlo a 'mean1'
mean1 = df['data1'].mean()

# Calcular el promedio de la columna 'data2' en el DataFrame 'df' y asignarlo a 'mean2'
mean2 = df['data2'].mean()

# Mostrar los valores promedio de 'data1' y 'data2'
print(mean1, mean2)
```
