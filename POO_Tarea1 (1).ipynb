```python
#BASE DE DATOS DE JUGADORES DE FUTBOL Y SUS GOLES POR 10 TEMPORADAS
import random


# Lista de nombres
nombres = ["Lewa", "Messi", "Ronaldo", "Neymar", "Suarez", "Aguero", "Ibra", "Haaland", "Mbappe", "Julio Cesar"]

# Generar matriz de datos aleatorios
matriz_datos = []
for _ in range(10):
    filas= [random.randint(15, 50) for _ in range(10)]  # 10 datos aleatorios por fila
    matriz_datos.append(filas)

```

```python
import pandas as pd
# Crear un DataFrame utilizando pandas
matriz = pd.DataFrame(matriz_datos, columns=[f"Temporada{i+1}" for i in range(10)], index=nombres)

# Mostrar la tabla
matriz
```
